```python
import numpy as np
import matplotlib.pyplot as plt

# Define the models
models = ['Linear SVM', 'Gaussian Naive Bayes', 'LSTMs', 'Random Forest', 'Deep Belief Network 1', 'Deep Belief Network 2', 'Neural Networks']

# Define the performance metrics for each model
roc_auc_scores = [0.6805555555555555, 0.5694444444444444, 0.521978021978022, 0.6703296703296703, 0.7417582417582418, 0.7417582417582418, 0.4375]
accuracy_scores = [0.6833333333333333, 0.5833333333333334, 0.4, 0.7, 0.75, 0.775, 0.45]
precision_scores = [0.5925925925925926, 0.48, 0.3611111111111111, 0.5714285714285714, 0.625, 0.6923076923076923, 0.3333333333333333]
recall_scores = [0.6666666666666666, 0.5, 0.9285714285714286, 0.5714285714285714, 0.7142857142857143, 0.6428571428571429, 0.375]
specificity_scores = [0.6944444444444444, 0.6388888888888888, 0.11538461538461539, 0.7692307692307693, 0.7692307692307693,0.8461538461538461, 0.5]
sensitivity_scores = recall_scores  # Sensitivity is the same as Recall in this context

# Set the width of the bars
bar_width = 0.14

# Set the positions of the bars on the x-axis
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]
r5 = [x + bar_width for x in r4]

# Create the grouped bar plot
plt.figure(figsize=(12, 8))
plt.bar(r1, roc_auc_scores, color='skyblue', width=bar_width, edgecolor='grey', label='ROC AUC')
plt.bar(r2, accuracy_scores, color='lightgreen', width=bar_width, edgecolor='grey', label='Accuracy')
plt.bar(r3, precision_scores, color='salmon', width=bar_width, edgecolor='grey', label='Precision')
plt.bar(r4, recall_scores, color='lightblue', width=bar_width, edgecolor='grey', label='Recall')
plt.bar(r5, specificity_scores, color='lightcoral', width=bar_width, edgecolor='grey', label='Specificity')

# Add xticks on the middle of the group bars
plt.xlabel('Models', fontweight='bold')
plt.xticks([r + 2 * bar_width for r in range(len(models))], models, rotation=45, ha='right')

# Add labels and title
plt.ylabel('Scores', fontweight='bold')
plt.title('Comparison of Model Performance Metrics', fontweight='bold')
plt.ylim(0, 1)  # Set y-axis limit for better visualization

# Add legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
